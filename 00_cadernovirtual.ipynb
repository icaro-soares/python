1. Comandos Iniciais e Exibição (print)
Para mostrar algo na tela, usamos a função print(). O conteúdo deve seguir regras específicas dependendo do tipo:

* Textos (Strings): Devem estar entre aspas (simples ou duplas).

* Números: São escritos sem aspas.

* O sinal de +: * Com números: Realiza a soma aritmética. Ex: print(7 + 4) → 11.

* Com strings: Realiza a concatenação (une os textos). Ex: print('7' + '4') → 74.

3. Tipos Primitivos
Os quatro tipos básicos fundamentais:
* int: Números inteiros (sem vírgula). Ex.: .7, -4, 0, 1975

* float: Números reais (com ponto flutuante). Ex.: 4.5, 0.076, -1.5

* bool: Valores lógicos (Booleanos); True, False

* str: Caracteres de texto (Strings). Ex.:'Olá', "Python"

4. Operações Aritméticas e Precedência
O Python segue uma ordem específica para resolver cálculos, muito similar à matemática:

* Parênteses ()

* Potência ** ou pow(x, y) (usando a função pow(x, y) a ordem de precedência é perdida)

* Multiplicação, Divisão, Divisão Inteira e Módulo *, /, //, %

* Adição e Subtração +, -

Exemplos de Operadores "Diferentes":
    // (Divisão inteira): Retorna apenas o número antes da vírgula.

    % (Módulo): Retorna o resto da divisão.

    ** (1/2): Forma de calcular a raiz quadrada. Ex: 81**(1/2) resulta em 9.0.

5. Formatação de Saída (F-Strings)
Existem formas modernas de exibir variáveis dentro de textos:

    Método .format(): print('A soma vale {}'.format(s))

    F-Strings (Mais prático): print(f'A soma vale {s}')

    Truques de Formatação:
    * {:>20}: Alinha à direita em 20 espaços.

    * {:^20}: Centraliza em 20 espaços.

    * {:=^20}: Centraliza colocando sinais de igual ao redor.

    * end=' ': Impede que o print pule para a próxima linha.

    * \n: Força a quebra de linha em qualquer lugar do texto.

## Análise e Fatiamento de Texto
O fatiamento permite pegar partes específicas de uma string usando índices. Lembre-se: o índice final não é incluído.

### Exemplos de Fatiamento:
* `frase[9]`: Pega o caractere na posição 9.
* `frase[9:21]`: Pega do 9 ao 20.
* `frase[9:21:2]`: Pega do 9 ao 20, pulando de 2 em 2.
* `frase[:5]`: Pega do 0 até o 4.
* `frase[5:]`: Pega do 5 até o fim.

```python
frase = 'Curso em Vídeo Python'

print(f'Comprimento: {len(frase)}') # Mostra o comprimento da string
print(f'Contar "o": {frase.count("o")}') # Conta o número de 'o'
print(f'Encontrar "deo": {frase.find("deo")}') # Verifica se o trecho se encontra na string
print(f'Existe "Curso"? {"Curso" in frase}') # Verifica se o trecho está presente

# Transformação
print(frase.replace('Python', 'Android')) # Troca 'Python' por 'Android', não altera a string origiral
print(frase.upper()) # Todas as letras maiúsculas
print(frase.lower()) # Todas as letras minúsculas
print(frase.capitalize()) # Só a primeira letra da frase
print(frase.title())      # Primeira letra de cada palavra
frase.rstrip() # Elimina os espaços em branco a direita da string
frase.lstrip() # Elimina os espaços em branco a esquerda da string
frase.strip() # Elimina os espaços em branco de ambos os lados da string
print(frase.split()) # Divide a string em pedaços menores levando em conta os espaços entre elas, recebe uma lista como retorno
frase.join('-') # Une as strings separadas anteriormente, recebe um argumento com o símbolo que será usado para fazer a união
print(frase.index('e')) # Retorna o índice da primeira aparição do elemento. Em caso de não existir retorna uma mensagem de erro.
print(frase.find('w')) # Faz quase a mesma coisa do .index, a diferença que quando não encontra o elemento retorna -1
frase.rindex('o') # Começa a procurar o elemento da direita para esquerda
frase.rfind('e') # Começa a procurar o elemento da direita para esquerda
print(frase.startswith('Curso')) # Verifica se a string começa com um termo, nos retornar True ou False
```

## Cores no Terminal (ANSI)
Para colorir o terminal, usamos o padrão `\033[style;text;backm`.

| Código | Significado |
| :--- | :--- |
| **Style** | 0 (None), 1 (Bold), 4 (Underline), 7 (Negative) |
| **Text** | 30 (Branco) a 37 (Cinza) |
| **Back** | 40 (Branco) a 47 (Cinza) |

```python
# Exemplo Prático de Cores
cores = {
    'limpa': '\033[m',
    'azul': '\033[34m',
    'amarelo': '\033[33m',
    'pretoebranco': '\033[7;30m'
}

print(f"{cores['azul']}Olá, Mundo!{cores['limpa']}")
print(f"{cores['pretoebranco']}Fundo invertido!{cores['limpa']}")
```
